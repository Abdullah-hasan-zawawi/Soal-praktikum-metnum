## Plot fungsi untuk melihat ada akar didalam intervalnya

Plot fungsi f(x) dalam interval yang diberikan

```python

def f(x):
    return ((exp(x)-exp(-x))/5)
interval=np.arange(-pi/5,pi/4,0.01)
hasil_plot=[]
%matplotlib inline
for i in interval:
    hasil_plot.append(f(i))
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.title('plot fungsi f(x)')
plt.plot(interval,hasil_plot,label='f(x)')
plt.legend()
```

### Bisection 

```python
def bisection(f,a,b,N,tol):
   
    if f(a)*f(b) >= 0:
        print("Bisection method fails.")
        return None
    a_n = a
    b_n = b
    list_iterasi_bisection=[]
    for n in range(1,N+1):
        m_n = (a_n + b_n)/2
        list_iterasi_bisection.append(m_n)
        f_m_n = f(m_n)
        if abs(f(m_n))<tol:
            break
        else:
            if f(a_n)*f_m_n < 0:
                    a_n = a_n
                    b_n = m_n
            elif f(b_n)*f_m_n < 0:
                    a_n = m_n
                    b_n = b_n
            elif f_m_n == 0:
                    pass

            else:
                    print("Bisection method fails.")

    return (list_iterasi_bisection)
batas_iterasi=input('sampai berapa iterasi yang dinginkan jika barisan tak konvergen ? :')
batas_iterasi=int(batas_iterasi)
toleransi=input('toleransi yang dinginkan pada program ? :')
toleransi=float(toleransi)
my_list=bisection(f,-pi/5,pi/4,batas_iterasi,tol=toleransi)
```

## Plot Barisan Akar

```python
data=pd.DataFrame(my_list,columns=['akar bisection '])
import matplotlib.pyplot as plt
hasil_f=[]
for i in my_list:
    hasil_f.append(f(i))
import seaborn as sns
plt.title('plot fungsi f(x) dengan menggunakan bisection')
plt.plot(interval,hasil_plot,label='f(x)')
plt.scatter(my_list,hasil_f,color='black')
plt.legend()
```

## Newton Method

```python
from numpy import sin,cos,log,exp    #definisiin fungsinya
def df(x,h=10**-12):                               #cari turunan fungsinya
    return (f(x+h)-f(x))/h

def Newton():
    p0=eval(input('masukan insial pointnya: '))
    N=eval(input('masukan jumlah batas iterasi yang diinginkan :'))
    tolerance=eval(input('masukan besar toleransinya :'))
    p = p0 - (f(p0)/df(p0))                                                    
    p0 = p
    hasil_fixed=[]
    for i in range(N):
        if abs(f(p))<tolerance:
            break
        else:
            p = p0 - (f(p0)/df(p0))                    #metode Newton
            p0 = p
            hasil_fixed.append(p0)
    return hasil_fixed
newton=Newton()
```
