```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from math import sin,cos,tan,exp
formula=input('masukan formula fungsinya :')
def f(x):
    return eval(formula)

low_bound = eval(input('Masukkan batas bawah integral : '))
up_bound = eval(input('Masukkan batas atas integral : '))
banyak_titik=eval(input('Masukan banyak titik : '))
banyak_titik=int(banyak_titik)
metode=input('masukan metode yang ingin digunakan simpson atau trapezoidal? ')



h_=up_bound-low_bound
step_size=h_/banyak_titik

list_x_array=np.arange(low_bound,up_bound,step_size)


def Trapezoid():                        #metode Trapezoid
    h = step_size
    sum_=[h/2*f(low_bound)]
    for i in range(1,banyak_titik-1):
        sum_.append(h*f(list_x_array[i]))

    return sum_
def Simpson():            
    h=step_size
    sum_=[h/3*f(low_bound)]
    for i in range(1,int(banyak_titik/2)-1):
        sum_.append((2/3)*h*f(list_x_array[2*i]))
    for i in range(1,int(banyak_titik/2)):
        sum_.append((4/3)*h*f(list_x_array[2*i-1]))
    sum_.append(h/3*f(up_bound))
    return sum_

if metode=='trapezoidal':
    hasil_trapezoidal=Trapezoid()
    
    hasil_fungsi=[]
    for i in list_x_array:
        hasil_fungsi.append(f(i))
    plt.plot(hasil_fungsi,label='fungsi sinus')
    plt.plot(hasil_trapezoidal,label='Trapezoidal')
    plt.legend()
elif metode=='simpson':
    hasil_simpson=Simpson()
    
    hasil_fungsi=[]
    for i in list_x_array:
        hasil_fungsi.append(f(i))
    plt.plot(hasil_fungsi,label='fungsi sinus')
    plt.plot(hasil_simpson,label='simpson')
    plt.legend()
else :
    print('kamu salah masukan nama metodenya')
```
